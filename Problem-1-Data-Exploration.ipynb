### 1.2 Basic Statistical Information

```python
def get_basic_stats(df):
    tdf = pd.DataFrame()
    tdf["Advertiser"] = np.sort(df.advertiser.unique())
    tdf["Cost"] = df.groupby("advertiser")["payprice"].sum().values
    tdf["Impressions"] = df.groupby("advertiser").size().values
    tdf["Clicks"] = df.groupby("advertiser")["click"].sum().values
    tdf["CTR"] = tdf["Clicks"] / tdf["Impressions"]
    tdf["CPM"] = (tdf["Cost"] / tdf["Impressions"] * 1000).round(2)
    tdf["eCPC"] = (tdf["Cost"] / tdf["Clicks"]).round(2)
    return tdf
```

```python
## Train Set Basic Statistical Information
tr_stats = get_basic_stats(tr_df)
tr_stats
```

```python
## Validation Set Basic Statistical Information
val_stats = get_basic_stats(val_df)
val_stats
```

### 2.2 CTR Analysis

```python
def get_group_ctr(df, attr):
    tdf = pd.DataFrame()
    tdf[attr] = np.sort(df[attr].unique())
    tdf["Impressions"] = df.groupby(attr).size().values
    tdf["Clicks"] = df.groupby(attr)["click"].sum().values
    tdf["CTR"] = tdf["Clicks"] / tdf["Impressions"]
    del tdf["Impressions"], tdf["Clicks"]
    return tdf
```

```python
def plot_group_ctr(df, attr, fsize, linestyle='o-'):
    tdf1 = get_group_ctr(df[df["advertiser"] == 1458], attr)
    tdf2 = get_group_ctr(df[df["advertiser"] == 3386], attr)
    fig = plt.figure(figsize=fsize, dpi=80)
    ax = fig.add_subplot(111)
    x1 = tdf1[attr].values
    x2 = tdf2[attr].values
    ax.plot([i for i in range(len(x1))], tdf1['CTR'].values, linestyle, color='darkcyan', label="1458")
    ax.plot([i for i in range(len(x2))], tdf2['CTR'].values, linestyle, color='firebrick', label="3386")
    ax.set_xticks([i for i in range(len(x1))])
    ax.set_xticklabels([str(i) for i in x1])
    plt.legend()
    plt.xlabel(attr)
    plt.ylabel("CTR")
    plt.show()
```

```python
plot_group_ctr(tr_df, "weekday", fsize=(8, 5))
```

```python
plot_group_ctr(tr_df, "hour", fsize=(12, 5))
```

```python
tr_df["os"] = tr_df["useragent"].apply(lambda x: x.split("_")[0])
plot_group_ctr(tr_df, "os", fsize=(8, 5), linestyle='o')
```

```python
tr_df["browser"] = tr_df["useragent"].apply(lambda x: x.split("_")[1])
plot_group_ctr(tr_df, "browser", fsize=(8, 5), linestyle='o')
```

```python
tr_df["tag_length"] = tr_df["usertag"].apply(len)
plot_group_ctr(tr_df, "tag_length", fsize=(16, 5))
```

```python
plot_group_ctr(tr_df, "region", fsize=(16, 5), linestyle='o')
```

## 3. Bidding Behaviour

```python
def plot_group_price(df, attr, fsize):
    target_df = pd.concat([df[df["advertiser"] == 1458], tr_df[tr_df["advertiser"] == 3386]])
    f, ax = plt.subplots(figsize=fsize)
    sns.factorplot(x=attr, y="payprice", hue="advertiser", data=target_df, ax=ax)
```

```python
paper_rc = {'lines.linewidth': 1, 'lines.markersize': 0.1}                  
sns.set_context("paper", rc = paper_rc)     
```

```python
plot_group_price(tr_df, "weekday", (8, 5))
```

```python
plot_group_price(tr_df, "hour", (16, 5))
```
